```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
1# Tải dữ liệu vào bộ khung dữ liệu dataframe
Train_data = pd.read_csv('/content/train.csv')
```

```python
#Đổi Cột type từ string sang int để quy tất cả các cột về dạng dữ liệu số 
Train_data['type'] = Train_data['type'].map({'white': 0,'red': 1})
```

```python
columns_need_to_drop = ['quality']
```

```python
#Tách Dữ Liệu Và cột được dán nhãn 
S = Train_data.drop(columns_need_to_drop,axis=1)
print(S)
```

```python
Y = Train_data['quality']
```

```python
S_train, S_test, Y_train, Y_test = train_test_split(S, Y, test_size= 0.2, random_state=1)
```
